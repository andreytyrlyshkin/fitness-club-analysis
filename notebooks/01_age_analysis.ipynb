## Анализ возраста посетителей
**Цель:** рассчитать средний возраст с фильтрацией и обработкой выбросов

```python
import numpy as np
```

```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

```python
#Task 1: расчёт среднего возраста по списку
avg_age_lst = np.mean(age_lst)
print(avg_age_lst)
```

```python
#Task 2: расчёт среднего возраста для посетителей моложе 50 лет
avg_age_lst = np.mean([i for i in age_lst if i < 50])
print(avg_age_lst)
```

```python
#Task 3 : функция, которая на вход принимает список и пороговое значение
def calc_avg_age(lst, threshold = 50):
    s = 0
    ctr = 0
    for i in lst:
        try:
            if i < threshold:
                s += float(i)
                ctr += 1
        except:
            print(f"Список содержит нечисловые значения")
    try:
        return round(s/ctr, 3)
    except ZeroDivisionError:
        print("В списке нет числовых значений, подходящих под порог")
            
```

```python
print(calc_avg_age(age_lst))
```

```python
#Task 4 : расчёт среднего значения для всех, за исключением лежащих вне интервала [Q1 - 1,5*IQR; Q3 + 1,5*IQR]
iqr = 1.5 * (np.percentile(age_lst, 75) - np.percentile(age_lst, 25))
advanced_avg_age_lst = np.mean([i for i in age_lst if (i > np.percentile(age_lst, 25) - iqr) & (i < np.percentile(age_lst, 75) + iqr)])
print(advanced_avg_age_lst)
```

```python
#Task 5 : функция, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы
def advanced_calc_avg_age(lst):
    try:
        iqr = 1.5 * (np.percentile(lst, 75) - np.percentile(lst, 25))
        s = 0
        ctr = 0
        for i in lst:
            if (i > np.percentile(lst, 25) - iqr) & (i < np.percentile(lst, 75) + iqr):
                s += i
                ctr += 1
    except:
        print("Список содержит нечисловые значения")
    try:
        return round(s/ctr, 3)
    except ZeroDivisionError:
        print("В списке не нашлось числовых значений")
```

```python
age_lst_test = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37, 110]
```

```python
sorted(age_lst_test)
```

```python
print(advanced_calc_avg_age(age_lst_test))
```

В тестовом списке видно, что возрасты 80, 88 и 110 оказались выбросами и не учлись в расчётах.
